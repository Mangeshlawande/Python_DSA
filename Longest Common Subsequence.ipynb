```python
def LCS(x,y,m,n):
    A = [[0 for x in range(n+1)] for x in range(m+1)]
    print('A-> ',A)
    for i in range(m+1):
        for j in range(n+1):
            print('i->',i,'j->',j)
            if i == 0 or j == 0:
                A[i][j] = 0
                print('i or j = 0 || A[i][j] = ',A[i][j],'\n')
            elif x[i-1] == y[j-1]:
                print('x = ',x[i-1],' y = ',y[j-1])
                A[i][j] = A[i-1][j-1] + 1
                print('x[i-1]=y[j-1] || A[i][j] = ',A[i][j],'\n')
            else:
                A[i][j] = max(A[i-1][j], A[i][j-1])
                print('x = ',x[i-1],' y = ',y[j-1])
                print('else part || A[i][j] = ',A[i][j],'\n')
                
    print('length = ',A[m][n])

    length = A[m][n]

    lcs = [None] * (length)
    print('lcs length ->',len(lcs),'\n')

    i = m
    j = n
    while i > 0 and j > 0:
        if x[i-1] == y[j-1]:
            lcs[length-1] = x[i-1]
            i = i - 1
            j = j - 1
            length = length - 1
            print('i=',i,'j=',j,'length=',length)
            print('lcs -> ',lcs,'\n')

        elif A[i-1][j] > A[i][j-1]:
            i = i - 1
            print('i decremented',i)
        else:
            j = j - 1
            print('j decremented',j)
            
    print('x = ',x,'\ny = ',y)
    print('Longest Common Subsequence =',lcs)


x = "ABCD"
y = "BCDE"
m = len(x)
n = len(y)
LCS(x,y,m,n)
```
