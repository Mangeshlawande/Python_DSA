# Try to compare & swap the adjacent element

```python
def BS(arr):
    for i in range(len(arr)):
    
     for j in range(0, len(arr)-i-1):
        
        if arr[j] > arr[j + 1]:
            
            temp = arr[j]
            arr[j] = arr[j+1]
            arr[j+1] = temp
```

```python
a = [6,5,8,4,2]
BS(a)
print("Sorted Array:")
print(a)
```

```python
def BS1(arr):
    for i in range(len(arr)):
        for j in range(0,len(arr)-i-1):
            
            print("i-->>",i,"j-->>",j,"\n")
            if arr[j] > arr[j+1]:
                
                print("Swap",arr[j],arr[j+1])
                
                
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
                
                print(arr,"\n")
                
            else:
                print("No Swap",arr[j],arr[j+1])
                print(arr,"\n")
                
```

```python
a = [6,5,4,8,2]
BS1(a)
print("Sorted Array:")
print(a)
```
