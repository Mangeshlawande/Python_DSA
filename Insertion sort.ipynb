```python
def IS(arr):  # ascending
    for i  in range (1,len(arr)):
        key = arr[i]
        j = i-1 
        print("Key-->>", key)
        print("j-->>",j)# return the index of array
        print("arr[j]-->>",arr[j])
        
        while j >=0 and key < arr[j]:
            print("array before copy -->",arr,'\n')
            print('Internal j value -->', j)
            arr[j + 1] = arr[j]
            j = j - 1 
            print("array before placing the key -->>",arr,'\n')
            
        print("key placed-->",key)
        arr[j + 1] = key
        print('array after placing the key -->',arr,'\n')
        
```

```python
a = [6,4,8,3,2]
IS(a)
print("Sorted Array")
print(a)
```

```python
a = [6,4,8,3,2]
IS1(a)
print("Sorted Array")
print(a)
```
