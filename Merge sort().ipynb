## merge sort()

```python
def MS(arr):
    if len(arr) > 1:
        print(1,'\n')
    
        mp = len(arr)//2  # // for floor
    
        left = arr[:mp]
    
        right = arr[mp:]
    
        print('left call',left)
        MS(left)
        print('right call',right)
        MS(right)
    
        i = j = 0
        k = 0
    
        print('left out-->',left)
        print('right out -->',right)
    
        while i < len(left) and j  < len(right):
        
            if left[i] < right[j]:
                print("i -->",i,"j-->",j)
                print(left[i],' left < right ',right[j])
            
                arr[k] = left[i]
            
                i = i + 1
                print('array left -->',arr)
                print('arr[k] -->',arr[k])
            
            else:
                print("i -->",i,"j-->",j)
            
                arr[k] = right[j]
            
                j = j + 1
                print('array right-->', arr)
            
            k = k + 1
            print('k-->',k)
        
        while i < len(left):
            print("i is-->",i,"arr[k]-->",arr[k])
        
            arr[k] = left[i]
            print("arr[k]-->",arr[k])
        
            i = i + 1
         
            k = k + 1
            print('left >> arr -->',arr,'k-->',k)
        
        while j < len(right):
            
             print("j is-->",j,"arr[k]-->",arr[k])
        
             arr[k] = right[j]
             print("arr[k]-->",arr[k])
             j = j + 1

             k = k + 1
             print('right >> arr-->',arr,'k-->',k)
        
def display(arr):
    
    for i in range (len(arr)):
        
        print('\n',arr[i])

        
```

```python
a = [9,1,5,3,8,2]
MS(a)
display(a)
```
